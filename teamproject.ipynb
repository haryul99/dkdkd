```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import matplotlib.pyplot as plt
import json
import re



response = requests.get("https://news.seoul.go.kr/traffic/archives/307")
html_doc = response.text

soup = BeautifulSoup(html_doc, "html.parser")
# print(soup)
soup.select('#post_content > table:nth-child(35) > tbody > tr:nth-child(3)')

timeuser_list = soup.select('#post_content > table:nth-child(35) > tbody > tr:nth-child(3)')

xpoint = []

# print(timeuser_list)

for n in timeuser_list :
    xpoint.append(n.get_text())

xpoint.pop(0)
    
my_array = np.array(xpoint)

xpoints = np.array(['2시','4시','6시','8시','10시','12시','14시','16시','18시','20시','22시'])
ypoints = np.array(['20대','30대','40대','50대','60대','70대','80대','','','',''])
zpoints = np.array(my_array)


with open("test3.txt", "w", encoding="utf-8") as phone :
    for n in xpoint :
        phone.write('%s'%n)
        

        
plt.rc('font',family='Malgun Gothic')

plt.plot(xpoints, ypoints,zpoints,'r+') #matplotlib라는 라이브러리에서 plot이라는 것으로 값을 넣음
plt.title('나이때별 시간의 확률')
plt.xlabel('시간대별 영업건수')
plt.ylabel('21월 말 기준 종사자 나이 ')

plt.show()
```
