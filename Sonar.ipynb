```python
from tensorflow.keras import models,Sequential, layers
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.layers import Dense
 
import pandas as pd
import numpy
import tensorflow as tf
  
# seed 값 설정
numpy.random.seed(3)
tf.random.set_seed(3)
  
# 데이터 입력
df = pd.read_csv('E:\source\TF\dataset\sonar.csv', header=None)

dataset = df.values
X = dataset[:,0:60].astype(float)
Y_obj = dataset[:,60]
  
# 문자열 변환
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
  
```

```python
print(Y)
```

```python
# 모델 설정
model = models.Sequential()
model.add(layers.Dense(24, input_dim=60, activation='relu')) #60개가 들어오는데 24개로 바꿔줌
model.add(layers.Dense(10, activation='relu')) #24개로 들어오는 걸 10개로 바꿔줌
model.add(layers.Dense(1, activation='sigmoid')) #10개로 들어오는 걸 1개로 (2진분류)

model.summary()
```

```python
# 모델 컴파일
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
  
# 모델 실행
model.fit(X, Y, epochs=200, batch_size=5)
  
# 결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(X, Y)[1]))
```
