```python
from tensorflow.keras import models, layers
import numpy as np
import tensorflow as tf
```

```python
np.random.seed(3)
tf.random.set_seed(3)
```

```python
Data_set = np.loadtxt("E:\source\TF\dataset\ThoraricSurgery.csv", delimiter=',')
```

```python
X = Data_set[:,0:17]
Y = Data_set[:,17]
```

```python
model_name = 'thoraric_surgery'
BATH_SIZE = 10
params = {'model_name' : model_name,
         'input_dimension' : 17,
         'hidden_dimension' : 30,
         'output_dimension' : 1}
```

```python
class ThoraricSurgery(tf.keras.Model): #반드시 케라스의 모델을 상속받아야함
    def __init__(self, **kargs):  #딕셔너리타입의 파라미터를 받아야함
        super(ThoraricSurgery, self).__init__(name=kargs['model_name']) #틀이 고정임
        self.fc1 = layers.Dense(kargs['hidden_dimension'], input_dim=kargs['input_dimension'], 
                               activation='relu')
        self.fc2 = layers.Dense(kargs['output_dimension'], activation='sigmoid')
        
    def call(self, x):
        x = self.fc1(x)
        x = self.fc2(x)
        return x
```

```python
model = ThoraricSurgery(**params)
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, Y, epochs=100, batch_size=10)
```

```python
model.build((BATH_SIZE, params['input_dimension'])) #수동빌드
```
